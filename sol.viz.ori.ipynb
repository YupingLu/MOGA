```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/data/ylu/11knobs/res/ori/0111_moga/gen_94_db.dat', header=None)
data = df.to_numpy()
da_t = data[:,15:17]
da = np.mean(da_t, axis=1)
ma = data[:,17]
```

```python
df1 = pd.read_csv('/data/ylu/11knobs/res/rand/rand_9/gen_63_db.dat', header=None)
data1 = df1.to_numpy()
da_t1 = data1[:,15:17]
da1 = np.mean(da_t1, axis=1)
ma1 = data1[:,17]
```

```python
df0 = pd.read_csv('/data/ylu/11knobs/res/sin/9_sin/gen_57_db.dat', header=None)
data0 = df0.to_numpy()
data3 = data0[data0[:,-2] == 1]
da_t0 = data3[:,15:17]
da0 = np.mean(da_t0, axis=1)
ma0 = data3[:,17]
```

```python
df4 = pd.read_csv('/data/ylu/11knobs/res/rd_track/rd_6.1_track/gen_1_db.dat', header=None)
data4 = df4.to_numpy()
data5 = data4[data4[:,-2] == 1]
da_t4 = data5[:,15:17]
da4 = np.mean(da_t4, axis=1)
ma4 = data5[:,17]
```

```python
df0 = pd.read_csv('/data/ylu/11knobs/res/track/8.1_track/gen_1_db.dat', header=None)
data0 = df0.to_numpy()
data3 = data0[data0[:,-2] == 1]
da_t0 = data3[:,15:17]
da0 = np.mean(da_t0, axis=1)
ma0 = data3[:,17]
```

```python
MEDIUM_SIZE = 20
#plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

plt.figure(figsize=(12,6))
#plt.ylim(-0.027, -0.0235) 
plt.scatter(abs(da), abs(ma)*100, color='b', marker=".", s=20, alpha=0.75, label='Tr-MOGA 1') 
plt.scatter(abs(da1), abs(ma1)*100, color='red', marker=".", s=20, alpha=0.75, label='Tr-MOGA 2')
plt.scatter(abs(da0), abs(ma0)*100, color='grey', marker=".", s=20, alpha=0.75, label='ML-MOGA 1')
plt.scatter(abs(da4), abs(ma4)*100, color='black', marker=".", s=20, alpha=0.75, label='ML-MOGA 2')
plt.title('MOGA Output Parameters')
plt.xlabel('Dynamic Aperture [a.u]')
plt.ylabel('Momentum Aperture [%]')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend(markerscale=4)
plt.grid()
plt.tight_layout()
plt.savefig('sol.pdf')
```
