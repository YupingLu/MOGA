```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def extract(fname):
    df = pd.read_csv(fname, header=None)
    data = df.to_numpy()
    da_t = data[:,15:17]
    da = np.mean(da_t, axis=1)
    ma = data[:,17]
    em = data[:,2]
    return da, ma, em
```

```python
da, ma, em = extract('/data/ylu/11knobs/res/ori/0111_moga/gen_94_db.dat') #tr-moga
da1, ma1, em1 = extract('/data/ylu/11knobs/res/rand/rand_9/gen_63_db.dat') #tr-moga both
#da2, ma2, em2 = extract('/data/ylu/11knobs/res/sin/9_sin/gen_57_db.dat') # moga seed
da2, ma2, em2 = extract('/data/ylu/11knobs/res/track/8.1_track/gen_1_db.dat') # tracking
da3, ma3, em3 = extract('/data/ylu/11knobs/res/rd_track/rd_6.1_track/gen_1_db.dat') # tracking both
```

```python
MEDIUM_SIZE = 20
#plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

plt.figure(figsize=(12,9))
plt.scatter(em, da, color='b', marker=".", alpha=0.75, label='Tr-MOGA')
plt.scatter(em2, da2, color='r', marker=".", alpha=0.75, label='rank-1 ML-MOGA')

plt.xlabel('Emittance [m rad]')
plt.ylabel('Total Diffusion Rate [a.u.]')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend(markerscale=4)
plt.tight_layout()
plt.savefig('fig_tr_vs_ml_da-eps.pdf')
```
