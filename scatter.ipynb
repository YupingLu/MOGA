```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def extract(fname):
    df = pd.read_csv(fname, header=None)
    data = df.to_numpy()
    da_t = data[:,15:17]
    da = np.mean(da_t, axis=1)
    ma = data[:,17]
    em = data[:,2]
    da.sort()
    ma.sort()
    em.sort()
    return da, ma, em
```

```python
da, ma, em = extract('/data/ylu/11knobs/res/rand/rand_9/gen_63_db.dat')
#da1, ma1, em1 = extract('/data/ylu/11knobs/res/rd_ml/rd_6_ml/gen_1150_db.dat')
da2, ma2, em2 = extract('/data/ylu/11knobs/res/rd_track/rd_6.1_track/gen_1_db.dat')
```

```python
x = list(range(1, 5001, 1))
```

```python
MEDIUM_SIZE = 20
#plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

plt.figure(figsize=(12,9))
plt.scatter(x, da, color='b', marker=".", alpha=0.75, label='Tr-MOGA')
#plt.scatter(x, da1, color='r', marker=".", alpha=0.75, label='ML-MOGA')
plt.scatter(x, da2, color='r', marker=".", alpha=0.75, label='ML-MOGA validated')

plt.xlabel('Sample No.')
plt.ylabel('Total Diffusion Rate [a.u.]')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend(markerscale=4)
plt.tight_layout()
plt.savefig('fig_sorted_objective_distributions_da.pdf')
```
