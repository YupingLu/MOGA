```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ori_gen/gen_50_db.dat', header=None)
data = df.to_numpy()
da_t = data[:,15:17]
da = np.mean(da_t, axis=1)
ma = data[:,17]
```

```python
df0 = pd.read_csv('0328_gen/gen_50_db.dat', header=None)
data0 = df0.to_numpy()
da_t0 = data0[:,15:17]
da0 = np.mean(da_t0, axis=1)
ma0 = data0[:,17]
```

```python
df1 = pd.read_csv('0330_gen/gen_50_db.dat', header=None)
data1 = df1.to_numpy()
da_t1 = data1[:,15:17]
da1 = np.mean(da_t1, axis=1)
ma1 = data1[:,17]
```

```python
df2 = pd.read_csv('0428_gen/gen_40_db.dat', header=None)
data2 = df2.to_numpy()
da_t2 = data2[:,15:17]
da2 = np.mean(da_t2, axis=1)
ma2 = data2[:,17]
```

```python
plt.figure(figsize=(12,8))
plt.ylim(-0.029, -0.026) 
plt.scatter(da, ma, color='b', marker=".", alpha=0.75, label='Blue: Original')  # arguments are passed to np.histogram
plt.scatter(da0, ma0, color='k', marker=".", alpha=0.75, label='Black: Random')
plt.scatter(da1, ma1, color='r', marker=".", alpha=0.75, label='Red: 10% Random')
plt.scatter(da2, ma2, color='y', marker=".", alpha=0.75, label='Yellow: Combined')
plt.title('MOGA')
plt.xlabel('DA')
plt.ylabel('MA')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend()
plt.grid()
plt.savefig('sol.jpg')
```
