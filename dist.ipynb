```python
import matplotlib.pyplot as plt
import pickle
```

```python
# load files
ods = pickle.load(open("ods11.pkl", "rb"))
x1, x2, x3 = ods["x1"], ods["x2"], ods["x3"]
a, b, c = ods["a"], ods["b"], ods["c"]
```

```python
MEDIUM_SIZE = 20
#plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

plt.figure(figsize=(12,6))
plt.scatter(x1, a, color='blue', marker=".", alpha=0.75, label='Tr-MOGA 1')
plt.scatter(x2, b, color='red', marker=".", alpha=0.75, label='Tr-MOGA 2')
#plt.scatter(x3, c, color='black', marker=".", alpha=0.75, label='ML-MOGA 1')
plt.xlabel('Generation Number')
plt.ylabel('Distance Metric $\delta_i$ [a.u.]')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend(markerscale=4)
#plt.grid()
plt.tight_layout()
plt.savefig('dist_input.pdf')
```
