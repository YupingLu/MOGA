```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("da.out", header=None, sep=' ')
```

```python
data = df.to_numpy()
train = data[:, -2]
test = data[:, -1]
x = data[:, 0]
```

```python
MEDIUM_SIZE = 20
#plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

plt.figure(figsize=(12,8))
plt.ylim(0, 2e+6) 

#plt.xlim(-0.027, -0.010) 
plt.plot(x, train, color='b', marker=".", alpha=0.75, label='Train')  
plt.plot(x, test, color='red', marker=".", alpha=0.75, label='Test')  
plt.title('Total Diffusion Rate [a.u.]')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend()

plt.tight_layout()
plt.savefig('train.pdf')
```
