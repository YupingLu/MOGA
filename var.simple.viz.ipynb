```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
date = '0328'
num = '50'
```

```python
df = pd.read_csv(date+'_gen/gen_'+num+'_db.dat', header=None)
```

```python
data = df.to_numpy()
X = data[:,20]
Y = data[:,21]
```

```python
plt.scatter(X, Y, color='b', marker=".", alpha=0.75)  # arguments are passed to np.histogram
plt.rcParams["figure.figsize"] = (6,6)
plt.title('MOGA Input Variables - gen '+num)
plt.xlabel('X')
plt.ylabel('Y')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.grid()
plt.savefig(date+'_'+num+'.jpg')
```
