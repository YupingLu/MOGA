```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('ori_gen/gen_50_db.dat', header=None)
data = df.to_numpy()
X = data[:,20]
Y = data[:,21]
```

```python
df0 = pd.read_csv('0328_gen/gen_50_db.dat', header=None)
data0 = df0.to_numpy()
X0 = data0[:,20]
Y0 = data0[:,21]
```

```python
df1 = pd.read_csv('0330_gen/gen_50_db.dat', header=None)
data1 = df1.to_numpy()
X1 = data1[:,20]
Y1 = data1[:,21]
```

```python
df2 = pd.read_csv('0428_gen/gen_20_db.dat', header=None)
data2 = df2.to_numpy()
X2 = data2[:,20]
Y2 = data2[:,21]
```

```python
plt.figure(figsize=(6,6))
#plt.ylim(-0.029, -0.026) 
plt.scatter(X, Y, color='b', marker=".", alpha=0.75, label='Blue: Original')  # arguments are passed to np.histogram
plt.scatter(X0, Y0, color='k', marker=".", alpha=0.75, label='Black: Random')
plt.scatter(X1, Y1, color='r', marker=".", alpha=0.75, label='Red: 10% Random')
plt.scatter(X2, Y2, color='y', marker=".", alpha=0.75, label='Yellow: Combined at Gen 20')
plt.title('MOGA Input Variables')
plt.xlabel('X')
plt.ylabel('Y')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
plt.legend()
plt.grid()
plt.savefig('var20.jpg')
```
