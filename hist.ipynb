```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
date = 'ori'
num = '1'
```

```python
df = pd.read_csv(date+'_gen/gen_'+num+'_db.dat', header=None)
```

```python
data = df.to_numpy()
da_t = data[:,15:17]
da = np.mean(da_t, axis=1)
ma = data[:,17]
```

```python
plt.figure(figsize=(9,6))
num_bins = 200
plt.hist(da, num_bins, facecolor='blue')
plt.title('MOGA DA - gen '+num)
plt.grid(True)
plt.savefig(date+'_'+num+'_da.jpg')
```

```python
plt.figure(figsize=(9,6))
num_bins = 200
plt.hist(ma, num_bins, facecolor='blue')
plt.title('MOGA MA - gen '+num)
plt.grid(True)
plt.savefig(date+'_'+num+'_ma.jpg')
```
