```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def extract(fname):
    df = pd.read_csv(fname, header=None)
    data = df.to_numpy()
    da_t = data[:,15:17]
    da = np.mean(da_t, axis=1)
    ma = data[:,17]
    em = data[:,2]
    return da, ma, em
```

```python
def extract1(fname):
    df = pd.read_csv(fname, header=None)
    data1 = df.to_numpy()
    data = data1[data1[:,-2] == 1]
    da_t = data[:,15:17]
    da = np.mean(da_t, axis=1)
    ma = data[:,17]
    em = data[:,2]
    return da, ma, em
```

```python
# tr-moga
da, ma, em = extract('/data/ylu/11knobs/res/ori/0111_moga/gen_94_db.dat')
# last gen of 1st ml iter
da1, ma1, em1 = extract('/data/ylu/11knobs/res/sin/9_sin/gen_57_db.dat')
# gen 1 of tr-moga
da2, ma2, em2 = extract('/data/ylu/11knobs/res/rand/rand_9/gen_63_db.dat')

min_da = min(min(da), min(da1), min(da2))
max_da = max(max(da), max(da1), max(da2))
min_ma = min(min(ma), min(ma1), min(ma2))
max_ma = max(max(ma), max(ma1), max(ma2))
min_em = min(min(em), min(em1), min(em2))
max_em = max(max(em), max(em1), max(em2))
```

```python
MEDIUM_SIZE = 20
#plt.rc('font', size=MEDIUM_SIZE) 
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
 
da, ma, em = extract('/data/ylu/11knobs/res/ori/1204_moga/gen_1_db.dat')
plt.figure(figsize=(12,8))
plt.ylim(min_ma, max_ma) 
plt.xlim(min_da, max_da)
plt.scatter(da, ma, c=em, cmap='plasma', vmin=min_em, vmax=max_em, marker=".", alpha=0.75)
plt.xlabel('Total Diffusion Rate [a.u.]')
plt.ylabel('-MA [ ]')
plt.tick_params(axis='both', direction='out', grid_color='gray', grid_alpha=0.5)
#plt.legend()
cb = plt.colorbar(label='Emittance [m rad]')
cb.ax.tick_params(labelsize=MEDIUM_SIZE)
plt.tight_layout()
plt.savefig('fig_tr-moga_gen1.pdf')
```
